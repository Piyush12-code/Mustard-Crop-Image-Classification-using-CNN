```python
# ============================================
# âœ… Setup & Mount Google Drive
# ============================================
from google.colab import drive
drive.mount('/content/drive')

!cp -r "/content/drive/MyDrive/Mustard dataset" "/content/"
!mv "/content/Mustard dataset" "/content/Mustard_dataset"

# ============================================
# âœ… Install & Import Libraries
# ============================================
!pip install -U fastai timm --quiet

from fastai.vision.all import *
import timm

# ============================================
# âœ… Define Path & Verify Structure
# ============================================
path = Path("/content/Mustard_dataset")

for root, dirs, files in os.walk(path):
    print(root, "->", len(files), "files")

assert (path/'train').exists() and (path/'test').exists(), "Train/Test folders missing!"

# ============================================
# âœ… Data Loaders with Strong Augmentations
# ============================================
# Improve data variability to reduce overfitting
dls = ImageDataLoaders.from_folder(
    path,
    train='train',
    valid='test',
    bs=64,                             # Larger batch size if GPU memory allows
    item_tfms=Resize(256),
    batch_tfms=aug_transforms(
        flip_vert=True,
        max_rotate=30,
        max_zoom=1.2,
        max_warp=0.2,
        max_lighting=0.3,
        p_affine=0.8,
        p_lighting=0.8
    )
)

print("Classes found:", dls.vocab)
dls.show_batch(max_n=8, figsize=(8,6))

# ============================================
# âœ… Model: Use a Stronger Architecture (ResNet50 / EfficientNet)
# ============================================
# Switch from ResNet34 â†’ EfficientNet or ResNet50 for better accuracy
learn = vision_learner(
    dls,
    'efficientnet_b0',     # lightweight and efficient model
    metrics=accuracy,
    pretrained=True
)

# Find the optimal learning rate
learn.lr_find()

# ============================================
# âœ… Fine-tuning Strategy
# ============================================
# 1. Train head first, then fine-tune entire model
learn.fine_tune(
    epochs=10,
    base_lr=3e-4,
    freeze_epochs=2,
    cbs=[
        SaveModelCallback(monitor='accuracy', fname='best_model'),
        EarlyStoppingCallback(monitor='accuracy', patience=3)
    ]
)

# ============================================
# âœ… Evaluate Model Performance
# ============================================
learn.show_results(max_n=8, figsize=(8,6))

interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix(figsize=(6,6))
interp.most_confused(min_val=5)

# ============================================
# âœ… Save & Export Model
# ============================================
learn.export("/content/mustard_fastai_efficientnet.pkl")
print("âœ… Model training complete and exported successfully.")
```

```python
# ===============================
# âœ… Print Classification Report and Accuracy Score
# ===============================
from sklearn.metrics import classification_report, accuracy_score
import torch

# Get predictions and actual labels for the validation set
# learn.get_preds returns a tuple: (predicted probabilities, actual labels)
preds, targets = learn.get_preds(dl=dls.valid)

# Get the predicted class index by finding the index of the maximum probability for each prediction
predicted_indices = torch.argmax(preds, dim=1)

# Convert actual and predicted indices to class names
actual_labels = [dls.vocab[t] for t in targets]
predicted_labels = [dls.vocab[p] for p in predicted_indices]

# Print classification report
print(classification_report(actual_labels, predicted_labels))

# Print accuracy score
print("accuracy : ", accuracy_score(actual_labels, predicted_labels))
```

```python
# ===============================================
# âœ… Batch Prediction using FastAI Exported Model
# ===============================================

from fastai.vision.all import *
import pandas as pd
import os

# -----------------------------
# 1ï¸âƒ£ Load the trained model
# -----------------------------
model_path = '/drive/MyDrive/mustard model/mustard_fastai_efficientnet.pkl'   # <-- change path if needed
learn_inf = load_learner(model_path)
print("âœ… Model loaded successfully!")

# -----------------------------
# 2ï¸âƒ£ Path to folder with test images
# -----------------------------
test_folder = '/drive/MyDrive/Mustard dataset/test'  # <-- put your test folder path here
image_files = get_image_files(test_folder)
print(f"ðŸ–¼ Found {len(image_files)} images for prediction.")

# -----------------------------
# 3ï¸âƒ£ Predict for each image
# -----------------------------
results = []

for img_path in image_files:
    try:
        pred, pred_idx, probs = learn_inf.predict(img_path)
        results.append({
            'image_name': os.path.basename(img_path),
            'predicted_class': str(pred),
            'confidence': float(probs[pred_idx])
        })
    except Exception as e:
        print(f"âš ï¸ Error processing {img_path}: {e}")

# -----------------------------
# 4ï¸âƒ£ Save results to CSV
# -----------------------------
df = pd.DataFrame(results)
output_csv = '/drive/MyDrive/Mustard dataset\prediction_results.csv'
df.to_csv(output_csv, index=False)

print(f"âœ… Predictions saved successfully to: {output_csv}")
```
